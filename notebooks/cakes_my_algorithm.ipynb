## 9. Relalization of model

* _K_-Nearest Neighbors (_k_-NN) is a supervised machine learning algorithm.
* In _k_-NN classification algorithm, the observed object will be classified according to the most common class in the set of the _k_ nearest objects (objects with the smallest distance to the observed object).

1. First we need to calculate distance. I will use Euclidean distance which is calculated like:
    $ EuclideanDistance = \sqrt{\sum \limits_{i = 1}^{n} (p_i - q_i)^2 } $, where $ P = (p_1, p_2 ... p_n) $ and $ Q = (q_1, q_2 ... q_n) $ are two points in N-dimensional space.



```python
def euclidean_distance(p, q):
    return np.sqrt(pow(p - q, 2).sum())
```

## 10. Calculating single prediction

#### Function parameters

* **_x_test_row_** - row which we are classifying.
* **_k_** - parameter k.
* **_x_train_set_** - we will classify **_x_test_row_** by calculating distance between it and every row in **_x_train_set_**.
* **_y_train_set_** - will be used for determining class based on.

#### Objects in function
* **_distances_** - dictionary with key - value pairs, key is index of row in **_x_train_set_**, value is euclidean distance between **_x_test_row_** (row which we are trying to classify) and all rows in **_x_train_set_** (we classify **_x_test_row_** based on values in this set).

* **_k_min_indexes_** - indexes in **_x_train_set_** of _k_ elements with smallest euclidean distance.
* **_k_min_classes_** - classes of elements on **_k_min_indexes_** positions.

```python
def get_single_prediction(x_test_row, k, x_train_set, y_train_set):
    distances = {}
    for index, row in enumerate(x_train_set.values):
        distances[index] = euclidean_distance(x_test_row, row)
    distances = sorted(distances.items(), key=lambda i: i[1])
    k_min_indexes = list(dict(distances[:k]).keys())
    k_min_classes = y_train_set.iloc[k_min_indexes].value_counts()
    return k_min_classes.idxmax()
```

```python
def predict(x_test_set, k, x_train_set, y_train_set):
    predictions = []
    for row in x_test_set:
        predictions.append(get_single_prediction(row, k, x_train_set, y_train_set))
    return predictions
```

## 11. Analysing results and evaluating performance of model

```python
prediction = predict(x_test.values, int(np.sqrt(data.shape[0])), x_train, y_train)

compare_df = pd.DataFrame(data=np.array([prediction, y_test]).transpose(), columns=['predicted', 'expected'])

count_wrong = compare_df['predicted'].compare(compare_df['expected']).shape[0]
count_right = compare_df.shape[0] - count_wrong

right_percentage = count_right / (count_right + count_wrong) * 100
wrong_percentage = count_wrong / (count_right + count_wrong) * 100

print("Score [RIGHT]: {:.2f} %".format(right_percentage))
print("Score [WRONG]: {:.2f} %".format(wrong_percentage))
```

## 12. Cross validation of parameter k

```python
k_max = int(np.sqrt(data.shape[0]))

best_k = k_max
best_score = 0

right_predictions_percentage = []

for k in range(1, k_max + 1):
    prediction = predict(x_test.values, k, x_train, y_train)
    compare_df = pd.DataFrame(data=np.array([prediction, y_test]).transpose(), columns=['predicted', 'expected'])
    count_wrong = compare_df['predicted'].compare(compare_df['expected']).shape[0]
    count_right = compare_df.shape[0] - count_wrong
    right_percentage = count_right / (count_right + count_wrong) * 100
    wrong_percentage = count_wrong / (count_right + count_wrong) * 100
    if right_percentage > best_score:
        best_score = right_percentage
        best_k = k
    right_predictions_percentage.append(right_percentage)

print(f'Best k is: {best_k}')
print(f'Best score is: {best_score}')

plt.scatter(np.linspace(1, k_max, num=len(right_predictions_percentage)), right_predictions_percentage, c='red')
plt.plot(np.linspace(1, k_max, num=len(right_predictions_percentage)), right_predictions_percentage)
plt.xlabel("k")
plt.ylabel("Score")
plt.show()
```
